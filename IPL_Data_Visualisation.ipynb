```python
#loading libraries
import numpy as np #numerical computing 
import pandas as pd #data processing, CSV file I/O
import matplotlib.pyplot as plt #visualization
import seaborn as sns #modern visualization
#setting theme for seaborn graphs
sns.set_style("darkgrid")
plt.rcParams['figure.figsize'] = (14, 8)
#reading input dataset
matches = pd.read_csv('matches.csv')
df=matches
class Node:
    def __init__(self, id=None,season=None,city=None,date=None,team1=None,team2=None,toss_winner=None,toss_decision=None,result=None,dl_applied=None,winner=None,win_by_runs=None,win_by_wickets=None,player_of_match=None,venue=None):
        self.id = id
        self.season = season
        self.city = city
        self.date = date
        self.team1 = team1
        self.team2 = team2
        self.toss_winner = toss_winner
        self.toss_decision = toss_decision
        self.result = result
        self.dl_applied = dl_applied
        self.winner = winner
        self.win_by_runs = win_by_runs
        self.win_by_wickets = win_by_wickets
        self.player_of_match = player_of_match
        self.venue = venue
        self.nextval = None

class SLinkedList:
    def __init__(self):
        self.headval = None

    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.id,printval.season,printval.city,printval.date,printval.team1,printval.team2,printval.toss_winner,printval.toss_decision,printval.result,printval.dl_applied,printval.winner,printval.win_by_runs,printval.win_by_wickets,printval.player_of_match,printval.venue)
            printval = printval.nextval
    
    def AtEnd(self, node):
        NewNode = node
        if self.headval is None:
            self.headval = NewNode
            return
        laste = self.headval
        while(laste.nextval):
            laste = laste.nextval
        laste.nextval=NewNode

    def size(self):
        current = self.headval
        count = 0
        while current:
            count += 1
            current = current.nextval
        #return count
        print("Size of the dataset:")
        print(count)
    def search(self, data):
        current = self.headval
        found = False

        while current and found is False:
            if current.nextval == data:
                found = True
                print("Data present")
                print(current)
            else:
                current = current.nextval
                if current is None:
                    raise ValueError("Data not in list")
        return current
list = SLinkedList()

for row in range(0,63):
    node = Node(df.iat[row,0],df.iat[row,1],df.iat[row,2],df.iat[row,3],df.iat[row,4],df.iat[row,5],df.iat[row,6],df.iat[row,7],df.iat[row,8],df.iat[row,9],df.iat[row,10],df.iat[row,11],df.iat[row,12],df.iat[row,13],df.iat[row,14])
    list.AtEnd(node)
def view():
    #print(matches)
    list.listprint()
def size():
    list.size()
def search():
    se=input("Enter the word to be searched for:")
    list.search(se)

def basic_info():
    print("Dimensions of the dataset:")
    print(matches.shape)
    print("Summary of the dataframe:")
    print(matches.info())
    print("Statistical Details:")
    print(matches.describe())  
def seasons():
    print("The following graph depicts the no. of matches every season")
    sns.countplot(x='season', data=matches)
    plt.show()
    print("Seasons in the dataset:")
    print(matches['season'].unique())
    print("No. of seasons in the dataset")
    print(len(matches['season'].unique()))
def successful_team():
    print("The following graph helps us identify the most successful team")
    data = matches.winner.value_counts()
    #print(data)
    sns.barplot(y = data.index, x = data, orient='h');  
def win():
    print("The team which won by maximum runs:")
    print(matches.iloc[matches['win_by_runs'].idxmax()]['winner'])
    print("The team which won by maximum wickets:")
    print(matches.iloc[matches['win_by_wickets'].idxmax()]['winner'])
    print("The team which won by minimum wickets:")
    print(matches.iloc[matches[matches['win_by_wickets'].ge(1)].win_by_wickets.idxmin()]['winner'])
    print("The team which won by closest margin(minimum runs):")
    print(matches.iloc[matches[matches['win_by_runs'].ge(1)].win_by_runs.idxmin()]['winner'])
def winf():
    print("The team which won by maximum runs:")
    print(matches.iloc[matches['win_by_runs'].idxmax()])
    print("The team which won by maximum wickets:")
    print(matches.iloc[matches['win_by_wickets'].idxmax()])
    print("The team which won by minimum wickets:")
    print(matches.iloc[matches[matches['win_by_wickets'].ge(1)].win_by_wickets.idxmin()])
    print("The team which won by closest margins(minimum runs):")
    print(matches.iloc[matches[matches['win_by_runs'].ge(1)].win_by_runs.idxmin()])
def top_player():
    print("The following graph helps us identify the most successful player")
    top_players = matches.player_of_match.value_counts()[:10]
    fig, ax = plt.subplots()
    ax.set_ylim([0,20])
    ax.set_ylabel("Count")
    ax.set_title("Top player of the match Winners")
    sns.barplot(x = top_players.index, y = top_players, orient='v'); #palette="Blues");
    plt.show()
def tossd():
    plt.figure(figsize=(12,4))
    sns.countplot(matches.toss_decision[matches.toss_winner == matches.winner])
    plt.show()
def heat():
    cor=matches.corr()
    plt.figure(figsize=(12,10))
    sns.heatmap(cor,annot=True,cmap="coolwarm")
    plt.show()
def dl():
    plt.figure(figsize=(20,9))
    sns.countplot(matches.city[matches.dl_applied==1])
    plt.show()
def venueb():
    plt.figure(figsize=(5,12))
    sns.swarmplot(x='win_by_wickets',y='venue',data=matches)
    plt.show()
    plt.figure(figsize=(5,12))
    sns.swarmplot(x='win_by_runs',y='venue',data=matches)
    plt.show()
def toss():
    ss = matches['toss_winner'] == matches['winner']
    print(ss.groupby(ss).size())
    sns.countplot(ss); 
    #venue_matches=matches.groupby('venue').count([['id']].sort_values(by='id',ascending=False)
def menu():
    print("""                              WELCOME!!
      This is a project based on the Indian Premier League.
      The following are the options you could use:
      1. View the dataset
      2. Size of the dataset
      3. Searching the dataset
      4. Get basic information about the data
      5. Seasons Information
      6. Most successful IPL team
      7. Winning teams
      8. Winning teams(Full details)
      9. Top players of match winners
      10. Has Toss winning helped in match winning?
      11. Does choosing batting or bowling first helped in match winning?
      12. Heatmap
      13. In which city does Weather affected matches?
      14. Best chasing venues
      15. Exit""")
    ch=int(input("Enter your choice:"))
    if(ch==1):
        view()
        menu()
    elif(ch==2):
        size()
        menu()
    elif(ch==3):
        search()
        menu()        
    elif(ch==4):
        basic_info()
        menu()
    elif(ch==5):
        seasons()
        menu()
    elif(ch==6):
        successful_team()
        menu()
    elif(ch==7):
        win()
        menu()
    elif(ch==8):
        winf()
        menu()
    elif(ch==9):
        top_player()
        menu()
    elif(ch==10):
        toss()
        menu()
    elif(ch==11):
        tossd()
        menu()
    elif(ch==12):
        heat()
        menu()
    elif(ch==13):
        dl()
        menu()
    elif(ch==14):
        venueb()
        menu()
    elif(ch==15):
        print("Thank You!")
    else:
        print("Enter a valid choice")
        menu()

menu()
```
